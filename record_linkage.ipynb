```python
import pandas as pd
import numpy as np

import googlemaps


import recordlinkage as rl
```

```python

# read in cleaned data sets


df1 = pd.read_csv('./data/permit_data_cleaned.csv', converters={'id': lambda x: str(x.strip()),
                                                                'zip': lambda x: str(x.strip()),
                                                                'add_zip': lambda x: str(x.strip()),
                                                                'city': lambda x: str(x.strip()),
                                                                'add': lambda x: str(x.strip())})



df2 = pd.read_csv('./data/well_data_cleaned.csv', converters={'permitno': lambda x: str(x.strip()),
                                                                'zip': lambda x: str(x.strip()),
                                                                'city': lambda x: str(x.strip()),
                                                                'add': lambda x: str(x.strip())})
```

```python
df2['zip'].value_counts()
```

```python
# join df2 to df1 on the following columns
# columns of interest are "add","city","zip","permitno","date","ar","ph" from df2
# the keys to join on in df1 are "add""
# keys to join on from df2 are "add"
# all columns from df2 should be joined to df1
# the output should be a dataframe with rows equal to the sum of the rows in df1 and df2
# the output should be saved as df3

df3 = pd.merge(df1, df2, left_on='id', right_on='permitno', how='outer')


```

```python
# check the number of valid entries in each column
df3
```

```python
# create a new dataframe keeping only rows that have a value for both "permitno" and "id"

df4 = df3.dropna(subset=['permitno', 'id'])

# check the number of valid entries in each column
df4

```

```python
# join df2 to df1 using recordlinkage library
# columns of interest are "add","city","zip","permitno","date","ar","ph" from df2
# the keys to join on in df1 are "id", "add", "city", "zip", and the 
# keys to join on from df2 are "permitno", "add", "city", "zip"
# all data from df2 should be joined to df1
# the output should be a dataframe with the same number of rows as df1
# and the columns from df1 and df2
# the output should be saved as df3


# create index
indexer = rl.Index()
indexer.block(left_on='city', right_on='city')
indexer.block(left_on='zip', right_on='zip')

# create candidate links
candidate_links = indexer.index(df1, df2)

# create a comparison object
c = rl.Compare()

# add the comparison functions

c.string('add', 'add', method='jarowinkler', threshold=0.85)
c.exact('id', 'permitno')

# create a feature vector
features = c.compute(candidate_links, df1, df2)
```

```python

# create a classifier
ecm = recordlinkage.ECMClassifier()

ecm.fit_predict(features)

# create a dataframe with the predicted links
df3 = pd.DataFrame(ecm.predict(features), columns=['match'])
```
